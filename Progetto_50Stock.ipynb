## 3 - For each stock in the dataset you should execute a rolling regression of logreturns as prescribed by the Market Model.

```python
import numpy as np

def logReturns(dfColumn):
  logret = np.log(dfColumn.astype('float64')/dfColumn.astype('float64').shift(1))
  return logret  

def logReturns180(dfColumn):
  logret = np.log(dfColumn.astype('float64')/dfColumn.astype('float64').shift(180))
  return logret  

SPdfRet = pd.DataFrame()

for column in SPdf:
  SPdfRet[column] = logReturns(SPdf[column])

SPdfRet = SPdfRet.iloc[1252:]



SPXdfRet = pd.DataFrame()

for column in SPXdf:
  SPXdfRet[column] = logReturns(SPXdf[column])

SPXdfRet = SPXdfRet.iloc[1252:]


SPXdfRet180 = pd.DataFrame()

for column in SPXdf:
  SPXdfRet180[column] = logReturns180(SPXdf[column])

SPXdfRet180 = SPXdfRet180.iloc[1252:]

SPXdfRet180
```

## 4 - The sample for the rolling regression must be 180 days. You will lose the first 180 days of the sample.

```python
import statsmodels.api as sm

betas = pd.DataFrame()
beta = pd.DataFrame()
alphas = pd.DataFrame()
alpha = pd.DataFrame()
alphasp = pd.DataFrame()
alphap = pd.DataFrame()
rsquares = pd.DataFrame()
rsquare = pd.DataFrame()
varResid = pd.DataFrame()
varResids = pd.DataFrame()
risks = pd.DataFrame()
risk = pd.DataFrame()
rs = pd.DataFrame()
r = pd.DataFrame()
delta_t = 180



length_period = SPdfRet.shape[0]

start = delta_t

for i in range(start, length_period, 7):
  print("-"*30)
  print(str(i)+" up to "+str(length_period))

  varMarket = np.var(SPXdfRet["SPX"][i-delta_t:i])

  print(SPdfRet.index[[i]])
  print(SPdfRet.index[[i]])
  print(SPdfRet.index[[i-delta_t]])

  for column in SPdfRet:
    # print(column)
    #beta_val_equity = beta_function(SPdfRet[column][i-delta_t:i-1], SPXdfRet["SPX"][i-delta_t:i-1])
    
    if(np.NAN not in SPdfRet[column][i-delta_t:i]):
      X = sm.add_constant(SPXdfRet["SPX"][i-delta_t:i])

      # print(X)

      mod = sm.OLS(SPdfRet[column][i-delta_t:i],X, )
      res = mod.fit()

    
    
      # print((res.resid))

      beta[column] = pd.Series([res.params.SPX], index = SPdfRet.index[[i]])
      alpha[column] = pd.Series([res.params.const], index = SPdfRet.index[[i]])
      alphap[column] = pd.Series([res.pvalues.const], index = SPdfRet.index[[i]])
      rsquare[column] = pd.Series([res.rsquared], index = SPdfRet.index[[i]])
      varResid[column] = pd.Series([np.var(res.resid)], index = SPdfRet.index[[i]])
      risk[column] = pd.Series([((res.params.SPX**2)*varMarket)+np.var(res.resid)], index = SPdfRet.index[[i]])
      r[column] = pd.Series([(res.params.const)+(res.params.SPX*SPXdfRet180["SPX"][i-delta_t]+np.var(res.resid))], index = SPdfRet.index[[i]])

    
    else:
      beta[column] = pd.Series([np.NAN], index = SPdfRet.index[[i]])
      alpha[column] = pd.Series([np.NAN], index = SPdfRet.index[[i]])
      alphap[column] = pd.Series([np.NAN], index = SPdfRet.index[[i]])
      rsquare[column] = pd.Series([np.NAN], index = SPdfRet.index[[i]])
      varResid[column] = pd.Series([np.NAN], index = SPdfRet.index[[i]])
      risk[column] = pd.Series([np.NAN], index = SPdfRet.index[[i]])
      r[column] = pd.Series([np.NAN], index = SPdfRet.index[[i]])

  betas = pd.concat([betas, beta], axis=0)
  alphas = pd.concat([alphas, alpha], axis=0)
  alphasp = pd.concat([alphasp, alphap], axis=0)
  rsquares = pd.concat([rsquares, rsquare], axis=0)
  varResids = pd.concat([varResids, varResid], axis=0)
  risks = pd.concat([risks, risk], axis=0)
  rs = pd.concat([rs, r], axis=0)
  beta = pd.DataFrame()
  alpha = pd.DataFrame()
  alphap = pd.DataFrame()
  rsquare = pd.DataFrame()
  varResid = pd.DataFrame()
  risk = pd.DataFrame()
  r = pd.DataFrame()
  
#betas
```
